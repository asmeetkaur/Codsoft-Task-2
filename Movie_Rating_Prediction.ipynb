Classify Movies Based on Rating [Excellent,Good and Average]

```python
data.columns
```

```python
def rating(rating):
  if rating>=7.0:
    return "Excellent"
  elif rating>=6.0:
    return "Good"
  else:
    return "Average"
data['rating_r']=data['Rating'].apply(rating)
data.head()
```

Find Unique Values From Genre

```python
data['Genre']
```

```python
list=[]
for value in data['Genre']:
  list.append(value.split(','))
list
```

```python
list1=[]
for item in list:
  for item1 in item:
    list1.append(item1)
list1
```

```python
list2=[]
for item in list1:
    if item not in list2:
      list2.append(item)
list2
```
